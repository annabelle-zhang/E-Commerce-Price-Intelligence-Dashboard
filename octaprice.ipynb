```python
# dataset taken from octaprice: https://github.com/octaprice/ecommerce-product-dataset
import pandas as pd
```

```python
df = pd.read_csv('octaprice_amazon_dataset.csv')
```

```python
# keeping only necessary columns

columns = ['brandName', 'breadcrumbs', 'color', 'currency', 'inStock',
       'listedPrice', 'material', 'mpn', 'name', 'new_path', 'nodeName', 'rating', 'reviewCount', 'salePrice',
       'scrapedDate', 'sku','url']
df = df[columns]
df.head()
```

```python
df['categories'] = df['nodeName'].str.split('>').str[0].str.strip()
```

```python
df['categories'].unique().tolist()
```

```python
electronics_categories = ['Amazon Devices & Accessories', 'Computers & Accessories', 'Electronics', 
                          'Cell Phones & Accessories', 'Software']

df = df[df['categories'].isin(electronics_categories)]
```

```python
df.info()
```

```python
df.to_csv('cleaned_octaprice.csv')
```
