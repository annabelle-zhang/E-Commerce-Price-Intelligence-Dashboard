```python
names = pd.read_csv('product_names.csv')
names['names'] = names['names'].str.split(',', n=1).str[0]
```

```python
names['names'] = names['names'].str.replace(r"[^a-zA-Z0-9\.\'\s\-]", '', regex=True)
names['names'] = names['names'].str.strip()
```

```python
names.to_csv('product_names.csv', index=False, header=['names'])
```
